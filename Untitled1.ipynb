```python
import random
import math
from tkinter import *

def etoile(L,d):
    R=L*(10**4)
    M=(d*(math.pi)/6)*(R**3)
    return (R,M)
    
def one_simal():
    taille=0
    taille = random.randint(1,30)

        
def simal_taille(n):
    list_size = list()
    for _ in range(n):
        list_size.append(one_simal())
    return list_size
        
def one_simal_pos_init(ray_min, ray_max):
    x = 0 #position sur l axe x
    y = 0 #position sur l axe y
    
    x = random.randint(ray_min, ray_max) * random.choice([-1, 1])
    y = random.randint(ray_min, ray_max) * random.choice([-1, 1])
        
    return (x,y)
        
def simal_pos_init(n, ray_min, ray_max):
    list_simal_pos = list()
    for _ in range(n):
        list_simal_pos.append(one_simal_pos_init(ray_min, ray_max))

        
def one_simal_masse(L):
    #3e12 #kg/km^3 densite des planetesimaux
    rayon = one_simal()
    return 3*10**12*(math.pi/6)*(L*10**4)**3
        
def simal_masse(n):
    list_simal_masse = list()
    for _ in range(n):
        list_simal_masse.append(one_simal_masse(3*10**4))
    
    return list_simal_masse
    
def simal_masse_taille(n):
    taille = simal_taille(n)
    masse = simal_masse(n)
    list_masse_taille = list()

    for i in range(n):
        list_masse_taille.append((taille[i], masse[i]))
        
    
def dict_simal(n, ray_min, ray_max):
    simal = dict()
    for i in range(1, n+1):
        simal[i] = ("taille:",one_simal(), "position:",one_simal_pos_init(ray_min, ray_max), "masse:",one_simal_masse(3*10^12))
            

import matplotlib.pyplot as plt

def one_simal_pos_time(y_min, y_max,time,v):
    x, y = one_simal_pos_init(y_min, y_max)
    if x > 0:
         angle_init = math.atan(y/x)
    else:
        angle_init = math.pi + math.atan(y/x)
        list_pos = list()
        list_pos.append(((x,y),angle_init))
        d = math.sqrt(x*x+y*y)
    
        for t in range(1,time+1):
            angle = angle_init + v*t
            x = math.cos(angle) * d
            y = math.sin(angle) * d
            list_pos.append(((x,y), angle))
        
        
        L1=[]
        L2=[]
    
        for ((x,y), angle) in list_pos:
            L1.append(x*10000000000)
            L2.append(y*10000000000)

    
        x=L1
        y=L2
    

        plt.scatter(x, y)
        plt.scatter([0],[0], s = 1000, c='orange')
        
        
def manysimal_pos_time(y_min, y_max,time, v , n):
 
    D=dict()
    for i in range(0,n):
        D[i]=one_simal_pos_time(y_min, y_max,time, v)
        

def collision(n, y_min, y_max, time, v,r):
    plusieurs_pos = manysimal_pos_time(n, y_min, y_max, time, v)
    masse_taille = simal_masse_taille(n)
    new_pos = list()

    col=dict()
    for i in range (0,n):
        col[i]=r
        
    n2=n
        
    for t in range (time):  
            
        for i in range(n2):    
            x_i, y_i = plusieurs_pos[t][i]
            masse_i, taille_i = masse_taille[i]
            
            for e in range(n2):
                x_n, y_n = plusieurs_pos[t][n2]
                masse_n ,taille_n = masse_taille[n2]
                    
                if x_i - x_n <= r and y_i - y_n <=r:
                    plusieurs_pos.remove[i]
                    plusieurs_pos.remove[e]
                    plusieurs_pos.append[((x_i + x_n)/2, (y_i + y_n)/2)]
                    
                    masse_taille.remove[i]
                    masse_taille.remove[e]
                    masse_taille.append[((masse_i+masse_n),(taille_i+taille_n))]  
                    n2=n2-1
                    col[n2]=r*2         

                    col.remove[i]
                    col.remove[e]
                    
                
                    
                    
            return n2


```
